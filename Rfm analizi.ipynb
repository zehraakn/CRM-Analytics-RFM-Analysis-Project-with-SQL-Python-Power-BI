```python
import pandas as pd
import datetime as dt

# Dosyayı oku (dosya yolunu kendine göre ayarla)
df = pd.read_csv("online_retail_data/online_retail_II.csv", encoding='ISO-8859-1')

# İlk 5 satırı göstererek veriyi incele
df.head()
```

```python
df = df[df['Invoice'].notnull()]         # Invoice boş değil
df = df[df['Customer ID'].notnull()]     # Müşteri ID boş değil
df = df[~df['Invoice'].astype(str).str.startswith('C')]  # İptal faturalar çıkar
df = df[df['Quantity'] > 0]              # Miktar pozitif
df = df[df['Price'] > 0]                 # Fiyat pozitif

# Toplam fiyat sütunu ekle
df['TotalPrice'] = df['Quantity'] * df['Price']

df.head()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Bugün tarihini veri içindeki son tarihten 1 gün sonraya ayarla
today_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

print(f"Bugünün tarihi: {today_date}")
```

```python
rfm = df.groupby('Customer ID').agg({
    'InvoiceDate': lambda x: (today_date - x.max()).days,  # Recency
    'Invoice': 'nunique',                                  # Frequency
    'TotalPrice': 'sum'                                    # Monetary
})

# Kolon isimlerini düzenle
rfm.columns = ['Recency', 'Frequency', 'Monetary']

# Monetary 0'dan büyük olanları al
rfm = rfm[rfm['Monetary'] > 0]

rfm.head()
```

```python
# Recency için ters sıralı kuantil (en küçük Recency = 5, en büyük = 1)
rfm['R'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])

# Frequency için kuantil (en sık alışveriş yapan 5 olur)
rfm['F'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])

# Monetary için kuantil (en çok harcayan 5 olur)
rfm['M'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])

# RFM skorlarını birleştir
rfm['RFM_SCORE'] = rfm['R'].astype(str) + rfm['F'].astype(str) + rfm['M'].astype(str)

rfm.head()
```

```python
# Segment map (R ve F kombinasyonları için isimlendirme)
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': "Can't Lose",
    r'3[1-2]': 'About to Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}

# R ve F skorlarını birleştir
rfm['Segment'] = rfm['R'].astype(str) + rfm['F'].astype(str)

# Regex ile segment isimlerini ata
rfm['Segment'] = rfm['Segment'].replace(seg_map, regex=True)

rfm.head()
```

```python
summary = rfm.groupby("Segment").agg({
    "Recency": ["min", "max", "mean", "sum", "count"],
    "Frequency": ["min", "max", "mean", "sum", "count"],
    "Monetary": ["min", "max", "mean", "sum", "count"]
}).round(2)

print(summary)
```

```python
# RFM ve Segment bilgilerini CSV'ye kaydet
rfm.reset_index().to_csv("rfm_segments.csv", index=False)

# Colab'dan indirmek için
from google.colab import files
files.download("rfm_segments.csv")
```
